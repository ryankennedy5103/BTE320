```python
def getValues(n: int) -> list:
    '''
    Creates a list of weights given by the user
    Input:
    - number of items in list
    Output:
    - a list of weights
    '''
    items = []
    for i in range(n):
        try:
            item = float(input("Enter weight value: "))
        except Exception as e:
            print(f'Error: {e}')
            item = float('nan')
        items.append(item)
    return items

def calcAverage(alist: list) -> float:
    '''
    Input:
    - a list of weights
    Output:
    - average value, excluding first two items
    '''
    try:
        return sum(alist[2:]) / len(alist[2:])
    except ZeroDivisionError:
        print('Error: List is empty')
        return float('nan')

def main():
    n = int(input("Enter number of weights: "))
    if n <= 0:
        raise ValueError("Number of weights must be greater than zero")
    weights = getValues(n)
    avg = calcAverage(weights)
    print(f'Average weight: {avg} lbs.')

main()
```
