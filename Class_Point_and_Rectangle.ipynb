```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self,other):
    return((self.x - other.x)**2+(self.y - other.y)**2)**.5
class Rectangle:
  def __init__(self)
    self.p1 = point(0,0)
    self.p2 = point(4,0)
    self.p3 = point(4,4)
    self.p4 = point(0,4)
    self.width = self.p1.distance(self.p2)
    self.height = self.p2.distance(self.p3)
  def area(self):
    return self.width * self.height
```
