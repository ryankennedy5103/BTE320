```python
# Parking garage
print("# of hours\tFee ($)")
print("--------------------")
for hours in range(1, 9):
   fee = 5 + hours * 2.5
   if fee < 10:
    fee = 10.0
    print(f"{hours}\t\t{fee}")
   elif fee > 20:
    fee = 20.0
    print(f"{hours}\t\t{fee}")
   else:
    print(f"{hours}\t\t{fee}")
```
