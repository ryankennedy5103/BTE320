```python
def calculate_fee(hours):
    fee = 5 + hours * 2.5
    if fee < 10:
        fee = 10.0
    elif fee > 20:
        fee = 20.0
    return fee

# Get input from user
hours = int(input("Enter number of hours: "))
price = calculate_fee(hours)
print(f"Fee for {hours} hours: ${price}")
```
