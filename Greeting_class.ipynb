```python
class Greeting:
    def __init__(self, first, last):
        self.first = first
        self.last = last
    def __str__(self):
        return 'Instance of class Greeting'
    def display(self, reverse=True):
        if reverse == True:
            print (f'{self.last}, {self.first}')
        else:
            print (f'{self.first} {self.last}')
name = Greeting('John', 'Doe')
print(name)
print(name.display())
print(name.display(reverse=False))
```
