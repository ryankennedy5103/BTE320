```python
def roi(cost: float, revenue: float) -> float:
    assert cost > 0, "The cost must be greater than 0"
    assert isinstance(cost and revenue, (float, int)), "The cost and revenue must be numbers"
    return (revenue-cost) / cost

if __name__ == "__main__":
    c = float(input("Enter the cost: "))
    r = float(input("Enter the revenue:"))
    retInvest=roi(c,r)
    print(f'The return on investment is {retInvest}')
```
