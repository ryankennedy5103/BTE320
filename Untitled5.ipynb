```python
class Student:
  def __init__(self, name, number):
    self._name = name
    self._number = number
    self._courses = []

  def get_name(self):
    return self._name

  def get_courses(self):
    return self._courses

  def set_name(self, newname):
    self._name = newname

  def __len__(self):
    return len(self._courses)

  def enroll(self, course):
    if course in self._courses:
      print('You have already enrolled in', course)
    else:
      self._courses.append(course)

s = Student('Bob',12345)
```

```python
s.get_name()
```

```python
s.get_courses()
```

```python
s.set_name('Bob Andrews')
```

```python
s.enroll('ABC123')
```
